```python
# the code we have the last module

import math
import pygame
pygame.init()

win = pygame.display.set_mode((500,500))
pygame.display.set_caption("FIRST GAME")

x = 50
y = 50
h = 60
w = 30
vel = 30
screen_width = 500
isJump = False
jumpCount = 10

run = True
while run:
    pygame.time.delay(10)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    pygame.draw.rect(win, (255,0,100), (x,y,h,w))
    pygame.display.update()

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and (x-vel) > 0:
        x -= vel
    if keys[pygame.K_RIGHT] and (x+h+vel) < screen_width:
        x += vel

    if not isJump:
        if keys[pygame.K_UP] and (y-vel) > 0:
            y -= vel
        if keys[pygame.K_DOWN] and (y+w+vel) < screen_width:
            y += vel

        if keys[pygame.K_SPACE]:
            isJump = True
    else:
        if jumpCount >= -10:
            neg = 1
            if jumpCount < 0:
                neg = -1
            y -= math.ceil((jumpCount ** 2) *0.5 * neg)
            jumpCount -= 1
        else:
            isJump = False 
            jumpCount = 10
            
    win.fill((0,0,0))
pygame.quit()
```
