```python
import pygame
pygame.init()
pygame.display.set_caption("First Game!")
win = pygame.display.set_mode((500,500))

x = 50
y = 50
height = 80
width = 60
vel = 5


run = True
while run:
    pygame.time.delay(100)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False


    # now we need to move that rectangle.
    # there are few ways that we can do that.
    # the first way is to use the event checking as above.

    # we can see if an key is pressed, then we can move the x and y coords of our rectangle.
    # but it wont be too great because it will only move our character just one time, not twice.

    # so what we need to do is that is to create a variable.
    keys = pygame.key.get_pressed()

    # now if we want to see which key is pressed we can just use an if statement,
    if keys[pygame.K_LEFT]:
        # now we want to move our character by the veocity in that direction
        # in pygame, the coordinates start from the top-left

        # if we want to move left we need to substract from the x coordinate
        x -= vel

    if keys[pygame.K_RIGHT]:
        x += vel

    if keys[pygame.K_UP]:
        y -= vel

    if keys[pygame.K_DOWN]:
        y += vel
    # these are arrow keys

    pygame.draw.rect(win, (0,255,0), (x,y,width,height))
    pygame.display.update()
        
pygame.quit()

# now if we run this we can see that it is pretty mess, 
# and instead of moving our character,it is creating a new character everytime.
```

```python
# so we need to fill the screen before we draw the rectangle.

import pygame
pygame.init()
pygame.display.set_caption("First Game!")
win = pygame.display.set_mode((500,500))

x = 50
y = 50
height = 80
width = 60
vel = 5


run = True
while run:
    pygame.time.delay(100)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False


    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT]:
        x -= vel

    if keys[pygame.K_RIGHT]:
        x += vel

    if keys[pygame.K_UP]:
        y -= vel

    if keys[pygame.K_DOWN]:
        y += vel

    # filling the screen with fill() method
    win.fill((0,0,0))
    # here it takes the color as an arguement

    pygame.draw.rect(win, (0,255,0), (x,y,width,height))
    pygame.display.update()

        
pygame.quit()
```
