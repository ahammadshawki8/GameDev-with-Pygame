```python
import pygame
pygame.init()

win = pygame.display.set_mode((500,500))
pygame.display.set_caption("FIRST GAME")

x = 50
y = 400
h = 30
w = 30
vel = 10
screen_width = 500

# for jumping we need to add to more variables
isJump = False
jumpCount = 10

run = True
while run:
    pygame.time.delay(100)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        pygame.draw.rect(win, (255,0,100), (x,y,h,w))
        pygame.display.update()
        
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and (x-vel) > 0:
            x -= vel
        if keys[pygame.K_RIGHT] and (x+w+vel) < screen_width:
            x += vel
            
        if not isJump:
            if keys[pygame.K_UP] and (y-vel) > 0:
                y -= vel
            if keys[pygame.K_DOWN] and (y+h+vel) < screen_width:
                y += vel

            if keys[pygame.K_SPACE]:
                isJump = True
        else:
            # here we will jump our character
            while isJump:
                if jumpCount >= -5:
                    # first we need to decrease our y variable by the sqaure of jumpCount
                    # since we have little space, so we stretch down the jump litlle bit
                    y -= (jumpCount ** 2) * 0.5
                    # now we need to decrease the jumpCount as things gradually slow down to reach the top
                    jumpCount -= 1
                else:
                    isJump = False  # after jump ends
                    jumpCount = 5  # reseting the var
            
        win.fill((0,0,0))
pygame.quit()
```
